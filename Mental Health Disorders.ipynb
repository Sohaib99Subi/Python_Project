```python
#Mental_health_disorder_Data

import pandas as pd
import numpy as np
import seaborn as sns




df = pd.read_csv(r"E:\Mental_health_disorder_Data.csv")
```

```python
#Select data from year 2004-2005


df2=df[df["Year"].between('2004','2005')]
df2.head(4)
```

```python
#Rename column

df2.rename({'Entity':'Country','Schizophrenia (%)':'Schizophrenia %','Bipolar disorder (%)':'Bipolar disorder %','Eating disorders (%)':'Eating disorders %','Anxiety disorders (%)':'Anxiety disorders %','Drug use disorders (%)':'Drug use disorders %','Depression (%)':'Depression %','Alcohol use disorders (%)':'Alcohol use disorders %'}, axis=1, inplace=True)
df2.head(1)
```

```python
#Checking datatypes

df2.dtypes
```

```python
#changing data types

df2['Schizophrenia %'] = pd.to_numeric(df2['Schizophrenia %'])

df2['Bipolar disorder %'] = pd.to_numeric(df2['Bipolar disorder %'])

df2['Eating disorders %'] = pd.to_numeric(df2['Eating disorders %'])
```

```python
#Drop unwanted rows

df2.drop([108455,108456,30557,30558,30585,30586,24285,24286,24313,24314,24341,24342,24369,24370] ,inplace = True)
```

```python
#Code for checking missing data

for col in df2.columns:
    pct_missing = np.mean(df2[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Filling Na with mean values

mean_value = df2.mean()
df2.fillna(mean_value, inplace = True)
```
